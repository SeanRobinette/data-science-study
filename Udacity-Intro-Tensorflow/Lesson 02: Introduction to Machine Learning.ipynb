```python
import tensorflow as tf
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
X    = np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
Y    = np.array([-40,  14, 32, 46, 59, 72, 100],  dtype=float)
for i in range(len(X)):
    print(str(X[i]) + " degrees Celsius = " + str(Y[i]) + " degrees Fahrenheit")
```

```python
L = [tf.keras.layers.Dense(units=1, input_shape=[1])]
model = tf.keras.Sequential(L)
```

```python
model.compile(tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')
```

```python
history = model.fit(X, Y, epochs=500, verbose=True)
```

```python
print(vars(history).keys())
nest_history = history.history
print(nest_history.keys())
plt.plot(nest_history['loss'])
```
